## Goal

###### This is a very simple project, made mostly to practice web scraping skills. The mains goal, besides the practice part, is to scrape prices from Mercado Livre, a brazilian website that works similarly as Amazon, to check for the better one to buy. From this simple process it's possible to develop a more complex project, as for example collecting data for a longer time and adding to a csv file to analyze the price evolution and understand the best moment to buy a certain product

```python
import requests
from bs4 import BeautifulSoup

url_base = 'https://lista.mercadolivre.com.br/'

product_name = input('Which product are you looking for? ')

page_number = 1  # starting in the first page
page_count = 0  # starts the page counting

while page_count < 3:  # limiting search
    # current URL
    url = url_base + product_name + f'_Desde_{(page_number-1)*50 + 1}'

    response = requests.get(url)
    cleanWeb = BeautifulSoup(response.text, 'html.parser')

    products = cleanWeb.findAll('div', attrs={'class': 'andes-card ui-search-result shops__cardStyles ui-search-result--core andes-card--flat andes-card--padding-16'})

    if not products:
        break  # if there's no more results, break the loop

    for product in products:
        title = product.find('h2', attrs={'class': 'ui-search-item__title shops__item-title'})
        link = product.find('a', attrs={'class': 'ui-search-item__group__element shops__items-group-details ui-search-link'})
        real = product.find('span', attrs={'class': 'andes-money-amount__fraction'})

        print('Product Title:', title.text)
        print('Product Link:', link['href'])
        print('Product Price: R$', real.text)
        print('\n')

    page_number += 1
    page_count += 1
```
